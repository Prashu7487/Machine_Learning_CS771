```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import pickle
```

### Loading dataset:

```python
# Load data from the pickle file
with open('data/mnist_small.pkl', 'rb') as file:
    loaded_data = pickle.load(file)

# Inspecting the loaded data
print(type(loaded_data))
print(loaded_data.keys())
# print(loaded_data['Y'])
# print(np.shape(loaded_data['Y']))
# print(loaded_data['X'])
```

```python
# Load images and labels in arraty
images = loaded_data['X']
labels=loaded_data['Y']

# Reduce dimensionality with PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(images)

# Reduce dimensionality with t-SNE
tsne = TSNE(n_components=2, perplexity=30, n_iter=300)
tsne_result = tsne.fit_transform(images)
```

```python
fig = plt.figure(figsize=(16,8))
plt.subplot(1, 2, 1)
sns.scatterplot(x=pca_result[:, 0].reshape(-1),y=pca_result[:, 1].reshape(-1), hue=labels.reshape(-1), palette=sns.color_palette())
plt.title('PCA Plot')
plt.subplot(1, 2, 2)
sns.scatterplot(x=tsne_result[:, 0].reshape(-1),y=tsne_result[:, 1].reshape(-1), hue=labels.reshape(-1), palette=sns.color_palette())
plt.title('tSNE Plot')
```

###  Observation: 
In comparison with PCA, tSNE projects datasets better and able to form better clusters with dataset (easily distinguishable) while PCA clusters are significantly overlapping so tSNE is better visulaization technique for large dimension datasets.
